```python
# Create new list of names to fix
snames_to_fix = []

# examine all names with 3 or more words
for sname in student_names:
    if len(sname.split()) >= 3: 
        snames_to_fix.append(sname)

# Get the number of students with names >= 3
len(snames_to_fix)
```

```python
prefixes = []
for sname in snames_to_fix:
    if len(sname.split()[0]) <= 4:
        prefixes.append(sname.split()[0])
        
print(prefixes)
```

```python
suffixes = []

for sname in snames_to_fix:
    if len(sname.split()[-1]) <=3:
        suffixes.append(sname.split()[-1])
        
print(suffixes)
```

```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
for fix in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(fix, "")
```
